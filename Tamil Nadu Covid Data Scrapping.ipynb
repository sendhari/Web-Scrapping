```python

District=[]
Total_Cases=[]
New_Cases=[]
Total_Deaths=[]
New_Deaths=[]
Total_Recovered=[]
Active_Cases=[]
Confirmed=[]
Active=[]
Recovered=[]
Deaths=[]

```

```python

pages = range(1,5)
url="https://coronaclusters.in/tamil-nadu"
results = requests.get(url, headers=headers)
soup = BeautifulSoup(results.text, "html.parser")
html_page = soup.find('tbody')##,attrs={'id':'district-data-table'})
test=html_page.prettify()

data_source = html_page.find_all('tr')
for a in data_source:

    
    district=a.find('th')
    District.append(district.text)
    specs = a.find('td')##,attrs={'class':'fMghEO'})
    for items in specs:
        detls =a.find_all('td')
        total_cases = a.find('td')        
        total_deaths=a.find('td')
        new_deaths=a.find('td')
        total_recovered=a.find('td')
        active_cases=a.find('td')
        Total_Cases.append(total_cases.text)
        New_Cases.append(detls[1].text)
        Total_Deaths.append(detls[2].text)
        New_Deaths.append(detls[3].text)
        Total_Recovered.append(detls[4].text)
        Active_Cases.append(detls[5].text)

```

```python
data1 = soup.find_all('h5')#, attrs={'class':'card-body'})
#print(data1)
Confirmed.append(data1[0].text)
Active.append(data1[1].text)
Recovered.append(data1[2].text)
Deaths.append(data1[3].text)
print(Confirmed)
```

```python
a = {'District':District,'Total_Cases':Total_Cases,'New Cases':New_Cases,'Total_Deaths':Total_Deaths,'New_Deaths':New_Deaths,'Total_Recovered':Total_Recovered,'Active_Cases':Active_Cases}
b = {'Total Confirmed Cases':Confirmed,'Active':Active,'Recovered':Recovered,'Deaths':Deaths}
```

```python
df1 = pd.DataFrame.from_dict(a, orient='index')
df1 = df1.transpose()
df2 = pd.DataFrame.from_dict(b, orient='index')
df2 = df2.transpose()
```
